```python
import cv2
import numpy as np
frame = cv2.imread(r"D:\projects\12.jpg")
frame = cv2.resize(frame,(800,448))
frame = cv2.circle(frame, (10, 170) , 5, (0, 0, 255) , -1) 
frame = cv2.circle(frame, (480, 130) , 5, (0, 0, 255) , -1)
frame = cv2.circle(frame, (790, 160) , 5, (0, 0, 255) , -1)
frame = cv2.circle(frame, (800, 448) , 5, (0, 0, 255) , -1) 
# pts1 = np.float32([[600, 1000], [600, 50], [1500, 500], [900, 50]]) 
pts1 = np.float32([[0, 170], [480, 130], [790, 160], [750, 400]])
pts2 = np.float32([[0, 0], [800, 0], [0, 448], [800, 448]])

# Apply Perspective Transform Algorithm 
matrix = cv2.getPerspectiveTransform(pts1, pts2) 
result = cv2.warpPerspective(frame, matrix, (800, 448)) 
# Wrap the transformed image 

cv2.imshow('frame', frame) # Inital Capture 
cv2.imshow('frame1', result) # Transformed Capture 

cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
# maping point in normal to perspective 
def Perspective(p,matrix):
    px = (matrix[0][0]*p[0] + matrix[0][1]*p[1] + matrix[0][2]) / ((matrix[2][0]*p[0] + matrix[2][1]*p[1] + matrix[2][2]))
    py = (matrix[1][0]*p[0] + matrix[1][1]*p[1] + matrix[1][2]) / ((matrix[2][0]*p[0] + matrix[2][1]*p[1] + matrix[2][2]))
    p_after = (int(px), int(py))
    return p_after
```
